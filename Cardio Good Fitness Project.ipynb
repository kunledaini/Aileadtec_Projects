Perform univariate and multivariate analyses

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# To visualize the Product using Catplot

sns.catplot(x='Product', data=df_cd, kind='count')
plt.title("Categorical Plot of Product")
plt.xlabel("Products - Threadmill")
plt.show()
```

- The above shows how the products are distributed abong the customers
- TM195 being used by 80 customers, TM498 by 60 and TM798 by 40

```python
# To visualize the column Age using histogram

sns.histplot(x='Age', data=df_cd)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.show()
```

- From the above, it shows that there are not many customers who fall between ages from 42 to 50 years of age
- We have more customers from 22 years of age to 36

```python
# Plotting a Multivarate

sns.boxplot(x='Product', y='Age', data=df_cd)
plt.title("Plot of Products vs Age")
plt.xlabel('Products')
plt.ylabel('Age')
plt.show()
```

- From above, the average age of customers who went for TM195 and TM498 which are almost the same is around 26 years of age
- TM798 having the average age around 27
- We can also see that the more older customers went for the first two products than they did for the last product
- Overall, fewer customers went for the TM798

```python
sns.histplot(binwidth=1,x='Product', hue='Gender', data=df_cd, stat="count", multiple="dodge")
plt.title("Plot of Products vs Gender")
plt.xlabel('Products')
plt.ylabel('Gender')
plt.show()
```

- From above, it shows that Male and Female went for the TM195 product almost equally.
- It is different with other 2 products
