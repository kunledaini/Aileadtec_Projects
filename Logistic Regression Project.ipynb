#### Data Visualization - Bivariate

```python
# Function to create scatter plot

def custom_scatterplot(x, y, title, data):
    plt.figure(figsize = (6, 4))
    sns.scatterplot(x=x, y=y, data=data)
    plt.title(title)
    plt.show()
```

##### Age Against Salary

```python
custom_scatterplot(x='age', y='salary', title = 'Plot of Age against Salary', data=df)
```

- Income does not seem to have effect on salary as indicated above

```python
sns.boxplot(data=df, x='salary', y='age')
plt.xlabel('salary')
plt.ylabel('age')
plt.title('Plot of Age against Salary')
plt.show()
```
